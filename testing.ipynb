```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import os
from tensorflow.keras.models import load_model
tf.get_logger().setLevel('FATAL')
```

```python
new_model = load_model('/mnt/d/DUMP/Notebook/BoneBroke/xray_analytics_cnn/models/imageclassifier.h5')

dir = '/mnt/d/DUMP/Notebook/BoneBroke/xray_analytics_cnn/data/val/fractured'

count = 0
count2 =0

for filename in os.listdir(dir):
    try:
        full_file_path = os.path.join(dir, filename)  # Create the full file path
        img = plt.imread(full_file_path)
        img_rgb = np.repeat(img, 3, axis=-1)
        resize = tf.image.resize(img, (256, 256))
        yhat = new_model.predict(np.expand_dims(resize/255, 0))
        count = count + 1
        print(count)
        print(filename)
        if yhat < 0.5:
            count2 = count2+1
            print(f'Predicted class is Fractured')
        else:
            print(f'Predicted class is Not Fractured')
    except Exception as e:
        print(f"Error processing image '{filename}': {str(e)}")
        
print("%d/%d"%(count2,count))
print(f"Processed {count} out of {len(os.listdir(dir))} images.")
    
```
